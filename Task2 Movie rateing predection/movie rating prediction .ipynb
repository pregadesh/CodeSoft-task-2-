```python
movie_data['Duration'] = movie_data['Duration'].astype(str)
movie_data['Year'] = movie_data['Year'].astype(str)
movie_data['Duration'] = movie_data['Duration'].str.extract('(\d+)').astype(float)
movie_data['Year'] = movie_data['Year'].str.extract('(\d+)').astype(float)
numerical_coloumns = ['Votes','Rating','Duration','Year']
correlation_matrix = movie_data[numerical_coloumns].corr()
plt.figure(figsize =(8,6))
sb.heatmap(correlation_matrix, annot = True, cmap ='coolwarm', vmin =-1, vmax = 1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
movie_data_sorted = movie_data.sort_values(by = 'Votes', ascending = False)
movie_data_sorted['Vote_count_Percentile'] = movie_data_sorted['Votes'].rank(pct = True) * 100
movie_data_sorted.reset_index(drop = True, inplace = True)
print(movie_data_sorted[['Name','Votes','Vote_count_Percentile']])
```

```python
movie_data = movie_data.dropna(subset = ['Votes'])
movie_data.head()
```

```python
x = movie_data[['Year','Duration','Votes']]
y = movie_data['Rating']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
rmse = mean_squared_error(y_test,y_pred,squared = False)
r2 = r2_score(y_test,y_pred)
print(f"mean absolute error:{mae}")
print(f"root mean squared error:{rmse}")
print(f"r-squared(R2) score:{r2}")
```

```python
y_test = np.random.rand(100) * 10
y_pred = np.random.rand(100) * 10
errors = y_test - y_pred
fig, axs = plt.subplots(3,1, figsize = (8,12))

axs[0].scatter(y_test, y_pred)
axs[0].set_xlabel('Actual Rating')
axs[0].set_ylabel('Predicted Rating')
axs[0].set_title('Actual vs Predicted Rating')


movie_sample = np.arange(1, len(y_pred) + 1)
axs[1].plot(movie_sample, y_pred, marker = 'o', linestyle = '-')
axs[1].set_xlabel("Movie sample")
axs[1].set_ylabel("Predicted Ratings")
axs[1].set_title("Predicted Ratings Across movie sample")
axs[1].tick_params(axis = 'x', rotation = 45)


axs[2].hist(errors, bins = 30)
axs[2].set_xlabel("Prediction error")
axs[2].set_ylabel("Frequency")
axs[2].set_title("Distribution of prediction error")
axs[2].axvline(x = 0, color = 'r', linestyle = '--')
plt.tight_layout()
plt.show()
```
